# Problem 1: Find the total number of crimes recorded in the CRIME table

# Problem 2: List community areas with per capita income less than 11000

```python
areas  = panda.read_sql_query("SELECT COMMUNITY_AREA_NAME,PER_CAPITA_INCOME FROM chicagocensusdata where PER_CAPITA_INCOME <11000 ",conn)
areas.head()
```

# Problem 3: List all case numbers for crimes involving minors? (children are not considered minors for the purposes of crime analysis)

```python
totalnumberofcrimes  = panda.read_sql_query("SELECT DISTINCT CASE_NUMBER, DESCRIPTION FROM chicagocrimedata WHERE DESCRIPTION LIKE '%MINOR%'",conn)
totalnumberofcrimes.head()
```

# Problem 4: List all kidnapping crimes involving a child?

```python
kidnapping  = panda.read_sql_query("SELECT DISTINCT CASE_NUMBER, PRIMARY_TYPE,DESCRIPTION FROM chicagocrimedata WHERE PRIMARY_TYPE='KIDNAPPING' and DESCRIPTION LIKE '%CHILD%'",conn)
kidnapping.head()
```

# Problem 5: What kinds of crimes were recorded at schools?

```python
school  = panda.read_sql_query("SELECT DISTINCT PRIMARY_TYPE FROM chicagocrimedata WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%'",conn)
school.head()
```

# Problem 6: List the average safety score for each type of school

```python
safety_score  = panda.read_sql_query("SELECT Elementary_Middle_High_School, AVG(SAFETY_SCORE) AVERAGE_SAFETY_SCORE FROM chicagopublicschools group by Elementary_Middle_High_School",conn)
safety_score.head()
```

# Problem 7: List 5 community areas with highest % of households below poverty line

```python
poverty  = panda.read_sql_query("SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY FROM chicagocensusdata ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5 ;",conn)
poverty.head()
```

# Problem 8: Which community area is most crime prone?

```python
crimeprone  = panda.read_sql_query("SELECT COMMUNITY_AREA_NUMBER, COUNT(COMMUNITY_AREA_NUMBER) FROM chicagocrimedata GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(COMMUNITY_AREA_NUMBER) DESC LIMIT 1",conn)
crimeprone.head()
```

# Problem 9: Use a sub-query to find the name of the community area with highest hardship index

```python
hardshipindex  = panda.read_sql_query("SELECT COMMUNITY_AREA_NAME, HARDSHIP_INDEX FROM  chicagocensusdata WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM chicagocensusdata)",conn)
hardshipindex.head()
```

# Problem 10: Use a sub-query to determine the Community Area Name with the most # of crimes?


```python
mostcrimes  = panda.read_sql_query("SELECT COMMUNITY_AREA_NAME  FROM chicagocensusdata WHERE COMMUNITY_AREA_NUMBER = (SELECT COMMUNITY_AREA_NUMBER FROM chicagocrimedata GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(COMMUNITY_AREA_NUMBER) DESC LIMIT 1) LIMIT 1",conn)
mostcrimes.head()
```
